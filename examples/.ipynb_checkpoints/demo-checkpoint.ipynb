# NotebookFlow demo

A four-node pipeline that loads inline data, filters it, summarises it, and prints the result. Open this notebook in JupyterLab or VS Code with the NotebookFlow extension installed, then run **NotebookFlow: Open Canvas** to see the graph render.

Every code cell below starts with a `# @node:` marker — that's what the parser uses to derive nodes. Edit the name in the marker (or double-click the node header in the canvas) and the change round-trips through the SyncEngine.

```python
# @node: Load Data  [input]  out=df
import pandas as pd

df = pd.DataFrame(
    {
        "id": [1, 2, 3, 4, 5, 6],
        "region": ["west", "east", "west", "north", "east", "north"],
        "revenue": [120.0, 87.5, None, 410.0, 56.0, 200.0],
    }
)
```

```python
# @node: Filter  [transform]  in=Load Data.df  out=clean_df
clean_df = df.dropna(subset=["revenue"]).copy()
clean_df["revenue"] = clean_df["revenue"].astype(float)
```

```python
# @node: Summarize  [transform]  in=Filter.clean_df  out=by_region
by_region = (
    clean_df.groupby("region", as_index=False)["revenue"]
    .agg(["sum", "mean", "count"])
    .rename(columns={"sum": "total", "mean": "avg", "count": "n"})
)
```

```python
# @node: Report  [output]  in=Summarize.by_region
print(by_region.to_string(index=False))
```
